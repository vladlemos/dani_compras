### Esse trecho abaixo é apenas para fazer ajustes em colunas para cruzamento de dados pelo nº de compra *



Apesar de funcionando, de acordo com a documentação do Pandas esse não é o melhor ou indicado modo de se fazer 


```python
# ano
planilha_requisicoes_sem_nulos['protocolo_r_ano'] = planilha_requisicoes_sem_nulos['Compra'].str.split('/').str[1]

# numero compra com 8 zeros
planilha_requisicoes_sem_nulos['pos'] = planilha_requisicoes_sem_nulos['Compra'].str.find('/')
planilha_requisicoes_sem_nulos['nc'] = planilha_requisicoes_sem_nulos.apply(lambda x: x['Compra'][0:x['pos']],axis=1)
planilha_requisicoes_sem_nulos['ncd'] = planilha_requisicoes_sem_nulos['nc'].str.zfill(8)

#numero da compra
planilha_requisicoes_sem_nulos['numero_compra'] = planilha_requisicoes_sem_nulos['protocolo_r_ano'] + planilha_requisicoes_sem_nulos['ncd']

#Apagando o que não vou usar
columns = ['pos', 'nc', 'protocolo_r_ano', 'ncd']
planilha_requisicoes_sem_nulos.drop(columns, inplace=True, axis=1)

# deixando os index como string
planilha_compras.Número = planilha_compras.Número.astype(str)
planilha_requisicoes_sem_nulos.numero_compra = planilha_requisicoes_sem_nulos.numero_compra.astype(str)

#renomeando coluna chave
planilha_compras = planilha_compras.rename(columns={'Número': 'numero_compra'})
```

### Alterado o nome da coluna chave Número na planilha de compras

## Merge de dados entre a planilha de compras e planilha de requisições pelo nº de compra

```python
mescla_compras_resquisicao = pd.merge(planilha_requisicoes_sem_nulos, planilha_compras, how = 'left', on = 'numero_compra')
```

```python
mescla_compras_resquisicao
```

## Gerando planilha com resultado mesclagem

```python
data = datetime.now().strftime("%d%m%Y-%H%M%S")
data
mescla_compras_resquisicao.to_excel('dados/merge-compras-requisicoes-{}.xlsx'.format(data), index = False)
```
